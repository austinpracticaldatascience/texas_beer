# 1. Data Cleaning

### A. Converting to dates to datetime datatype

```python
def timestamp(df,col):
    
    # convert to datetime
    df[col] = pd.to_datetime(df[col])

    # Updated info
    print(df.info())
    
timestamp(df,"Responsibility Begin Date")
timestamp(df,"Responsibility End Date")
timestamp(df,"Obligation End Date")
```

### B. Sort rows by "Obligation End Date" Descending

```python
df = df.sort_values(by=['Obligation End Date'])
df.tail(4)
```

### D. Check for Null Values

```python
def null_sum():

    # Check if there are any null values
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    x = missing.head()
    
    return x
    
null_sum()
```

## 2. Create a subset of Austin Mixed Beverages data

```python
# Focus on Austin
df_aus = df[df['Taxpayer County'] == 227]
```

```python
def shape_inspect(df):
    
    # Print dataframe shape
    print("The shape of the DataFrame is:", df.shape)
    
    # Skip a line
    print("\n")

    # Inspect data
    x = df.head(4)
    
    return x

shape_inspect(df_aus)
```

```python
aus_month = df_aus.groupby(['TABC Permit Number','Taxpayer Number','Location Name','Location Address','Obligation End Date']).agg('sum')
```

```python
aus_month = aus_month[['Liquor Receipts','Wine Receipts','Beer Receipts','Total Receipts']]

print(aus_month.shape)
```

```python
aus_month['Liquor %'] = aus_month['Liquor Receipts'] / aus_month['Total Receipts']
```

```python
aus_month['Wine %'] = aus_month['Wine Receipts'] / aus_month['Total Receipts']
```

```python
aus_month['Beer %'] = aus_month['Beer Receipts'] / aus_month['Total Receipts']
```

```python
aus_month = aus_month.sort_values(by=['Location Name','Obligation End Date']).reset_index()

aus_month.head(4)
```

```python
def joint(df,name,add):

    print("Mean percentage of sales per month that are beer is", (aus_month.loc[aus_month['Location Name'].str.contains(name, case=False) & aus_month['Location Address'].str.contains(add, case=False)]['Beer %'].mean())*100, "%.")

    t = aus_month.loc[aus_month['Location Name'].str.contains(name, case=False) & aus_month['Location Address'].str.contains(add, case=False)]
    
    return t
```

```python
def joint_tabc(df,tabc,name,add):

    print("Mean percentage of sales per month that are beer is", (aus_month.loc[aus_month['TABC Permit Number'].str.contains(tabc, case=False) & aus_month['Location Name'].str.contains(name, case=False) & aus_month['Location Address'].str.contains(add, case=False)]['Beer %'].mean())*100, "%.")

    t = aus_month.loc[aus_month['TABC Permit Number'].str.contains(tabc, case=False) & aus_month['Location Name'].str.contains(name, case=False) & aus_month['Location Address'].str.contains(add, case=False)].tail(12)
    
    return t
```

```python
easy_tiger_east = joint_tabc(aus_month,'mb','easy','1501')

easy_tiger_east
```

```python
easy_tiger_north = joint_tabc(aus_month,'mb','easy','int')

easy_tiger_north
```

```python
easy_tiger_south = joint_tabc(aus_month,'mb','easy','3508')

easy_tiger_south
```

```python
import matplotlib.pyplot as plt

# X axis
x = easy_tiger_south['Obligation End Date'].tail(12)

y_south = easy_tiger_south['Beer %'].tail(12)
y_east = easy_tiger_east['Beer %'].tail(12)
y_north = easy_tiger_north['Beer %'].tail(12)

# Plot lines
plt.plot(x, y_south, label='South')
plt.plot(x, y_north, label='North')
plt.plot(x, y_east, label='East')
plt.title("Percentage of Receipts that are beer \n at Easy Tiger locations Receipts")
# Set ylabel to be horizontal
plt.ylabel("Percent", labelpad=-25, loc='top', rotation='horizontal', fontsize=8)
plt.xlabel("Last 12 months", fontsize=8)
plt.legend()
plt.show()
```
